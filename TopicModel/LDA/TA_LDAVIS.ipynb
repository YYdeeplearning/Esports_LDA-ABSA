```python
import joblib

from gensim.models import CoherenceModel

from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pyLDAvis

from itertools import chain

import time
```

```python
year = 2014

corpus = joblib.load('./outputs/lda/ldamp_' + str(year) + '_corpus.pkl')
id2word = joblib.load('./outputs/lda/ldamp_' + str(year) + '_id2word.pkl')
model_list = joblib.load('./data/lda/ldamp_' + str(year) + '.pkl')

dat = joblib.load('./data/lda/sec_dt_PMDA_F' + str(year) + '.pkl')
dat = dat['PMDA']
```

```python
# Wordcloud
lda = model_list[20]
print(lda.num_topics)
lda_corpus = lda[corpus]
wc = WordCloud(background_color="black", stopwords='december')  #, max_words=1000, mask=alice_mask)
for t in range(lda.num_topics):
    plt.figure()
    plt.imshow(wc.fit_words(dict(lda.show_topic(t, 30))))
    plt.axis("off")
    plt.title("Topic #" + str(t))
    plt.show()
```
